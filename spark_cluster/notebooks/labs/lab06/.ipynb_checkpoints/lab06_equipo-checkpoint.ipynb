```python
from equipo.spark_utils import SparkUtils
agencies_schema = SparkUtils.generate_schema([
    ("show_id", "string"), 
    ("type", "string"),
    ("title", "string"),
    ("director", "string"),
    ("country", "string"),
    ("rating", "string"),
    ("duration", "string"),
    ("listed_in", "string"),
    ("release_year", "integer"),
    ("date_added", "date")
])

agencies_df = spark.read \
                .schema(agencies_schema) \
                .option("header", "true") \
                .csv("/home/jovyan/notebooks/data/netflix_dataset/netflix1.csv")

agencies_df.printSchema()

agencies_df.show(5, truncate=False)
```
